## Making our word embeded by pretrained embedding model 'GloVe6B'  
 - our word from keras news20 data
 - but this data is dirty. need preprocessing..
 - finally, predict 'random sentence' category.

```python
data_path = keras.utils.get_file(
    "news20.tar.gz",
    "http://www.cs.cmu.edu/afs/cs.cmu.edu/project/theo-20/www/data/news20.tar.gz",
    untar=True,
)
```

```python
import os
import pathlib

data_dir = pathlib.Path(data_path).parent / "20_newsgroup" # PosixPath('/root/.keras/datasets/20_newsgroup')
dirnames = os.listdir(data_dir)
print("Number of directories:", len(dirnames))
print("Directory names:", dirnames)
```

```python
fnames = os.listdir(data_dir / "comp.graphics")
print("Number of files in comp.graphics:", len(fnames))
print("Some example filenames:", fnames[:5])
print("\n")
print(open(data_dir / "comp.graphics" / "38764").read())
```

```python
# delete header because Newgroup has category.. ex. Newsgroups ~ Lines ... but it is not exactly in ten lines...
samples = []
labels = []
class_names = []
class_index = 0
for dirname in sorted(os.listdir(data_dir)):
    class_names.append(dirname)
    dirpath = data_dir / dirname
    fnames = os.listdir(dirpath)
    print("Processing %s, %d files found" % (dirname, len(fnames)))
    for fname in fnames:
        fpath = dirpath / fname
        f = open(fpath, encoding="latin-1")
        content = f.read()
        lines = content.split("\n")
        lines = lines[10:]           # May not header is not in 10 lines...
        content = "\n".join(lines)
        samples.append(content)
        labels.append(class_index)
    class_index += 1

print("Classes:", class_names)
print("Number of samples:", len(samples))
```

```python
class_names
```

```python
# Shuffle the data
seed = 1337
rng = np.random.RandomState(seed)
rng.shuffle(samples)
rng = np.random.RandomState(seed)
rng.shuffle(labels)

# Extract a training & validation split
validation_split = 0.2
num_validation_samples = int(validation_split * len(samples))
train_samples = samples[:-num_validation_samples] # 15998 samples
val_samples = samples[-num_validation_samples:] # 3999 samples
train_labels = labels[:-num_validation_samples] # 20 int labels about news category ex.[1,19,0,4,...]
val_labels = labels[-num_validation_samples:]
```

```python
from tensorflow.keras.layers.experimental.preprocessing import TextVectorization

# create vectorization layer
vectorizer = TextVectorization(max_tokens=20000, output_sequence_length=200) # <tensorflow.python.keras.layers.preprocessing.text_vectorization.TextVectorization at 0x7f90e60cac18>
text_ds = tf.data.Dataset.from_tensor_slices(train_samples).batch(128) # <BatchDataset shapes: (None,), types: tf.string>
vectorizer.adapt(text_ds)
```

```python
voc = vectorizer.get_vocabulary()
word_index = dict(zip(voc, range(len(voc)))) # {'':0, '[UNK]:1, 'the':2 ... }
```

```python
voc[:10] # it has 20000 word. frequencer word is in 20000. voc[0]=padding, voc[1] = out of voca. so Unknown
```

```python
# load pretrained embedding model
!wget http://nlp.stanford.edu/data/glove.6B.zip
!unzip -q glove.6B.zip
```

```python
pre_embeddings_index = {} # { 'word' : array([0.123, ... 0.234],dtype=float32), ... }
with open(os.path.join('glove.6B.100d.txt'), encoding='utf-8') as f:
    for line in f:
        word, coefs = line.split(maxsplit=1) # coefs is like [0.1 0.7 0.9 ...]
        coefs = np.fromstring(coefs, "f", sep=" ") # coefs is like 0.1 0.7 0.9 ...
        pre_embeddings_index[word] = coefs

print("Found %s word vectors." % len(pre_embeddings_index))
```

```python
# our word convert pretrained model's embedding.
num_tokens = len(voc) + 2
embedding_dim = 100
hits = 0
misses = 0

# Prepare embedding matrix
embedding_matrix = np.zeros((num_tokens, embedding_dim))
for word, i in word_index.items(): # get word in our news dict
    embedding_vector = pre_embeddings_index.get(word) # is our word in pretrained embedding dict?
    if embedding_vector is not None:
        # Words not found in embedding index will be all-zeros.
        # This includes the representation for "padding" and "OOV"
        embedding_matrix[i] = embedding_vector # our word embedded by pretrained model. embedding_matrix is like array([[0., 0., ...*100],..., [0.1, 0.2 *100]])
        hits += 1
    else:
        misses += 1
print("Converted %d words (%d misses)" % (hits, misses)) #our 2052 word is not in pretrained embedding models. 
```

```python
from tensorflow.keras.layers import Embedding

embedding_layer = Embedding(
    num_tokens,
    embedding_dim,
    embeddings_initializer=keras.initializers.Constant(embedding_matrix), ## very important
    trainable=False
)
```

```python
from tensorflow.keras import layers

int_sequences_input = keras.Input(shape=(None,), dtype="int64")
embedded_sequences = embedding_layer(int_sequences_input)
x = layers.Conv1D(128, 5, activation="relu")(embedded_sequences)
x = layers.MaxPooling1D(5)(x)
x = layers.Conv1D(128, 5, activation="relu")(x)
x = layers.MaxPooling1D(5)(x)
x = layers.Conv1D(128, 5, activation="relu")(x)
x = layers.GlobalMaxPooling1D()(x)
x = layers.Dense(128, activation="relu")(x)
x = layers.Dropout(0.5)(x)
preds = layers.Dense(len(class_names), activation="softmax")(x)
model = keras.Model(int_sequences_input, preds)
model.summary()
```

```python
x_train = vectorizer(np.array([[s] for s in train_samples])).numpy()
x_val = vectorizer(np.array([[s] for s in val_samples])).numpy()

y_train = np.array(train_labels)
y_val = np.array(val_labels)
```

```python
# train
model.compile(
    loss="sparse_categorical_crossentropy", optimizer="Adam", metrics=["acc"]
)
model.fit(x_train, y_train, batch_size=128, epochs=20, validation_data=(x_val, y_val))
```

```python
import tensorflow_hub as hub
import tensorflow as tf
from keras import backend as K
import urllib.request
import pandas as pd
import numpy as np
```

```python
# Download ELMo.
elmo = hub.Module("https://tfhub.dev/google/elmo/1", trainable=True)
```

```python
sess = tf.Session()
K.set_session(sess)
sess.run(tf.global_variables_initializer())
sess.run(tf.tables_initializer())
```

```python
# data
urllib.request.urlretrieve("https://raw.githubusercontent.com/mohitgupta-omg/Kaggle-SMS-Spam-Collection-Dataset-/master/spam.csv", filename="spam.csv")
data = pd.read_csv('spam.csv', encoding='latin-1')
data[:5]
```

```python
# label Integer encoding.
data['v1'] = data['v1'].replace(['ham','spam'],[0,1])
```

```python
# Store X and y separately.,
y_data = list(data['v1'])
X_data = list(data['v2'])
```

```python
# Train & Testing split
n_of_train = int(len(X_data) * 0.8)
n_of_test = int(len(X_data) - n_of_train)
X_train = np.asarray(X_data[:n_of_train]) 
y_train = np.asarray(y_data[:n_of_train]) 
X_test = np.asarray(X_data[n_of_train:]) 
y_test = np.asarray(y_data[n_of_train:]) 
```

```python
# Conversion required for use with Keras.
def ELMoEmbedding(x):
    return elmo(tf.squeeze(tf.cast(x, tf.string)), as_dict=True, signature="default")["default"]
```

```python
# Modeling
from keras.models import Model
from keras.layers import Dense, Lambda, Input
```

```python
# Define the model.
input_text = Input(shape=(1,), dtype=tf.string)
embedding_layer = Lambda(ELMoEmbedding, output_shape=(1024, ))(input_text)               # Lambda wraps up as a layer.
hidden_layer = Dense(256, activation='relu')(embedding_layer)
output_layer = Dense(1, activation='sigmoid')(hidden_layer)
my_model = Model(inputs=[input_text], outputs=output_layer)
my_model.compile(loss='binary_crossentropy', optimizer='adam', metrics=['accuracy'])
```

```python
# Training. 
# Notice that epochs = 1!!!
history = my_model.fit(X_train, y_train, epochs=1, batch_size=60)
```

```python
# Test for accuracy.
print("\n Test accuracy : %.4f" % (my_model.evaluate(X_test, y_test)[1]))
```
