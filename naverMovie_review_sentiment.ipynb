## Import

```
import pandas as pd
import numpy as np
import re
import urllib.request
from konlpy.tag import Okt,Hannanum,Kkma,Komoran,Mecab # so many tokenization method
from tensorflow.keras.preprocessing.text import Tokenizer
from tensorflow.keras.preprocessing.sequence import pad_sequences
```

## Data load

```
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_train.txt", filename="ratings_train.txt")
urllib.request.urlretrieve("https://raw.githubusercontent.com/e9t/nsmc/master/ratings_test.txt", filename="ratings_test.txt")
train_data = pd.read_table('ratings_train.txt')
test_data = pd.read_table('ratings_test.txt')
```

```
print('훈련용 리뷰 개수 :',len(train_data))
```

```
print('테스트용 리뷰 개수 :',len(test_data))
```

```
test_data[:5]
```

## Preprocessing

```
train_data.drop_duplicates(subset=['document'], inplace=True) # document 열에서 중복인 내용이 있다면 중복 제거
train_data['document'] = train_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") # 한글과 공백을 제외하고 모두 제거
train_data['document'].replace('', np.nan, inplace=True) # 빈 값을 null값으로 만들고,
train_data = train_data.dropna(how = 'any') # null값이 있는 행 제거

test_data.drop_duplicates(subset = ['document'], inplace=True) 
test_data['document'] = test_data['document'].str.replace("[^ㄱ-ㅎㅏ-ㅣ가-힣 ]","") 
test_data['document'].replace('', np.nan, inplace=True) 
test_data = test_data.dropna(how='any')

print('전처리 후 트레인용 샘플의 개수 :',len(train_data))
print('전처리 후 테스트용 샘플의 개수 :',len(test_data))
```

## Tokenization, Stopword
 - 토큰화 참고 https://konlpy.org/en/latest/api/konlpy.tag/

```
stopwords = ['의','가','이','은','들','는','좀','잘','걍','과','도','를','으로','자','에','와','한','하다']
```

```
okt = Okt()
mecab = Mecab()
komoran = Komoran()
kkma = Kkma()
hannanum = Hannanum()
```

```
X_train = []
for sentence in train_data['document']:
    temp_X = []
    temp_X = mecab.morphs(sentence) # 토큰화 ex.mecab.morphs(sentence),komoran.morphs(sentence) 
    temp_X = [word for word in temp_X if not word in stopwords] # 불용어 제거
    X_train.append(temp_X)
```

```
X_test = []
for sentence in test_data['document']:
    temp_X = []
    temp_X = mecab.morphs(sentence) 
    temp_X = [word for word in temp_X if not word in stopwords] 
    X_test.append(temp_X)
```

## Integer encoding
 - set vocab_size, threshold
 - ignore rare count word
 - delete empty sentence ( because some sentences are composed only rare word) 

```
vocab_size = 0
threshold = 3 # 빈도수 3번 이하 단어는 버림.
```

```
tokenizer = Tokenizer()
tokenizer.fit_on_texts(X_train)
```

```
total_cnt = len(tokenizer.word_index) # 단어의 수
rare_cnt = 0 # 등장 빈도수가 threshold보다 작은 단어의 개수를 카운트
total_freq = 0 # 훈련 데이터의 전체 단어 빈도수 총 합
rare_freq = 0 # 등장 빈도수가 threshold보다 작은 단어의 등장 빈도수의 총 합

# 단어와 빈도수의 쌍(pair)을 key와 value로 받는다.
for key, value in tokenizer.word_counts.items():
    total_freq = total_freq + value

    # 단어의 등장 빈도수가 threshold보다 작으면
    if(value < threshold):
        rare_cnt = rare_cnt + 1
        rare_freq = rare_freq + value
  
print('단어 집합(vocabulary)의 크기 :',total_cnt)
print('등장 빈도가 %s번 이하인 희귀 단어의 수: %s'%(threshold - 1, rare_cnt))
print("단어 집합에서 희귀 단어의 비율:", (rare_cnt / total_cnt)*100)
print("전체 등장 빈도에서 희귀 단어 등장 빈도 비율:", (rare_freq / total_freq)*100)
```

```
vocab_size = total_cnt - rare_cnt + 2 # 0번 패딩 토큰과 1번 OOV 토큰을 고려하여 +2
tokenizer = Tokenizer(vocab_size, oov_token = 'OOV') #OOV means out ot vocabulary. 즉, 사전에 없는 단어는 어떻게 인코딩 할지.
tokenizer.fit_on_texts(X_train)
X_train = tokenizer.texts_to_sequences(X_train) #이 부분이 정수인코딩하는 부분임.
X_test = tokenizer.texts_to_sequences(X_test)

print('단어 집합의 크기 :',vocab_size)
print(X_train[:3])
```

```
y_train = np.array(train_data['label'])
y_test = np.array(test_data['label'])
drop_train = [index for index, sentence in enumerate(X_train) if len(sentence) < 1] #find empty index
X_train = np.delete(X_train, drop_train, axis=0)
y_train = np.delete(y_train, drop_train, axis=0)

print(len(X_train))
print(len(y_train))
```

## Padding
 - set max_len

```
print('리뷰의 최대 길이 :',max(len(l) for l in X_train))
print('리뷰의 평균 길이 :',sum(map(len, X_train))/len(X_train))
```

```
def below_threshold_len(max_len, nested_list):
  cnt = 0
  for s in nested_list:
    if(len(s) <= max_len):
        cnt += 1
  print('전체 샘플 중 길이가 %s 이하인 샘플의 비율: %s'%(max_len, (cnt / len(nested_list))*100))
max_len = 50
below_threshold_len(max_len, X_train)
```

```
X_train_pd = pad_sequences(X_train, maxlen = max_len)
X_test_pd = pad_sequences(X_test, maxlen = max_len)

print(len(X_train_pd))
print(len(X_test_pd))
```

## Embedding & Modeling & Train
 - Embdding method is "tensorflow.keras.layers.Embedding()"
 - model is "LSTM". set unit count "128"
 - "Early Stoppin" & "Check point"is used in training
 - optimizer is "rmsprop"
 - epoch is "15", batch_size is "60"
 - it has about 0.85 val acc. good luck!

```
from tensorflow.keras.layers import Embedding         # so many Embedding
from tensorflow.keras.layers import Dense, GRU, LSTM, SimpleRNN, Bidirectional, Dropout, Conv1D, GlobalMaxPooling1D  # so many Model
from tensorflow.keras.models import Sequential, load_model
from tensorflow.keras.callbacks import EarlyStopping, ModelCheckpoint
```

```
model = Sequential()
model.add(Embedding(vocab_size, 100))
model.add(LSTM(128)) #or GRU(128) or SimpleRNN(128) or Bidirectional(LSTM(128)) or model.add(Conv1D(100, 3, padding='valid', activation='relu')) #3 is kernel size
# if use Conv1D
# model.add(GlobalMaxPooling1D())
# model.add(Dense(128, activation='relu'))
# model.add(Dropout(0.5))
model.add(Dense(1, activation='sigmoid'))
```

```
es = EarlyStopping(monitor='val_loss', mode='min', verbose=1, patience=4) #val_loss가 4번 연속 증가하면 학습 중단
mc = ModelCheckpoint('best_model.h5', monitor='val_acc', mode='max', verbose=1, save_best_only=True) #val_acc가 이전보다 증가해야 모델 저장
epochs = 15
batch_size = 60
```

```
model.compile(optimizer='rmsprop', loss='binary_crossentropy', metrics=['acc'])
history = model.fit(X_train_pd, y_train, epochs=epochs, callbacks=[es, mc], batch_size=batch_size, validation_split=0.2)
```

```
loaded_model = load_model('best_model.h5')
print("\n 테스트 정확도: %.4f" % (loaded_model.evaluate(X_test_pd, y_test)[1]))
```

## Predict

```
def sentiment_predict(new_sentence):
  new_sentence = mecab.morphs(new_sentence) # 토큰화
  new_sentence = [word for word in new_sentence if not word in stopwords] # 불용어 제거
  encoded = tokenizer.texts_to_sequences([new_sentence]) # 정수 인코딩
  pad_new = pad_sequences(encoded, maxlen = max_len) # 패딩
  score = float(loaded_model.predict(pad_new)) # 예측
  if(score > 0.5):
    print("{:.2f}% 확률로 긍정 리뷰입니다.\n".format(score * 100))
  else:
    print("{:.2f}% 확률로 부정 리뷰입니다.\n".format((1 - score) * 100))
```
